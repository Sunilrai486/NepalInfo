```python
import os
import re
import torch
import pandas as pd
from transformers import Trainer, TrainingArguments
from transformers import GPT2LMHeadModel, GPT2Tokenizer, GPT2Config
from transformers import TextDataset, DataCollatorForLanguageModeling
```

```python
# Loading the text data from the file.
with open("NepalInfo.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Initializing the tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Tokenizing the text.
input_ids = tokenizer.encode(text, return_tensors="pt")
```

```python
# Check if the 'content' folder exists, if not create it.
if not os.path.exists('/content'):
    os.mkdir('/content')
```

```python
# Defining the dataset and data collator
dataset = TextDataset(tokenizer=tokenizer, file_path="NepalInfo.txt", block_size=128)
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

```python
# Defining the training arguments for the Trainer model.
training_args = TrainingArguments(
    output_dir="/content",
    overwrite_output_dir=True,
    num_train_epochs=10,
    per_device_train_batch_size=16,
    save_steps=5_000,
    save_total_limit=3,
    learning_rate=2e-5,
    weight_decay=0.01,
)
```

```python
# Creating the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=dataset,
)

# Training the model.
trainer.train()
```

```python
# Saving the trained model.
model.save_pretrained("/content")
```

```python
# Loading fine-tuned model and tokenizer.
model_dir = "/content"
model = GPT2LMHeadModel.from_pretrained(model_dir)
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
# Defining a function to generate content and answer question based on the input.
def answer_question(prompt, max_length=80, words_per_line=20):
    # Tokenizing the input.
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generating the content.
    attention_mask = torch.ones(input_ids.shape, dtype=torch.long)
    output = model.generate(input_ids, max_length=max_length, num_return_sequences=1, attention_mask = attention_mask)
    generated_texts = tokenizer.decode(output[0], skip_special_tokens=True)

    # Splitting generated text into sentences
    sentences = re.split(r'(?<=[.,])\s+', generated_texts)

    # Removing duplicate sentences
    unique_sentences = []
    observed_sentences = set()
    for sentence in sentences:
        if sentence not in observed_sentences:
            observed_sentences.add(sentence)
            unique_sentences.append(sentence)
    processed_texts = ' '.join(unique_sentences)

    # Splitting the processed texts into chunks.
    words = processed_texts.split()
    lines = [' '.join(words[i:i+words_per_line]) for i in range(0, len(words), words_per_line)]

    return '\n'.join(lines)
```

```python
# Interactive UI to get information about Nepal.
while True:
    question = input("""\nGet Nepal Info:
    Introduction
    Land
    People
    Economy
    Government and society
    Cultural life
    History
    (type 'q' to quit): """)
    if question.lower() == "q":
        break
    answer = answer_question(question)
    print("\nAnswer:\n")
    print(answer + "\n")
```
